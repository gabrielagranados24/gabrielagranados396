```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar el archivo CSV
archivo_csv = 'ventasropa.csv'  # Cambia la ruta según tu archivo
df = pd.read_csv(archivo_csv)

# Conteo de productos más vendidos
productos_mas_vendidos = df['Item Purchased'].value_counts().head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=productos_mas_vendidos.index, y=productos_mas_vendidos.values, palette='viridis')
plt.title('Los 10 Productos Más Vendidos')
plt.xlabel('Producto')
plt.ylabel('Cantidad Vendida')
plt.xticks(rotation=45, ha='right')  # Para rotar etiquetas
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar el archivo CSV

df = pd.read_csv('ventasropa.csv')

# Conteo de métodos de pago
metodos_pago = df['Payment Method'].value_counts()

plt.figure(figsize=(8, 6))
sns.barplot(x=metodos_pago.index, y=metodos_pago.values, palette='coolwarm')
plt.title('Ventas por Método de Pago')
plt.xlabel('Método de Pago')
plt.ylabel('Cantidad de Ventas')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar archivo CSV
archivo_csv = 'ventasropa.csv'  # Cambia la ruta según tu archivo
df = pd.read_csv(archivo_csv)

# Agrupar datos para los 10 productos más vendidos
categorias = df.groupby("Item Purchased")["Purchase Amount (USD)"].sum().nlargest(10)

# Crear gráfico de barras
plt.figure(figsize=(12, 6))
plt.bar(categorias.index, categorias.values, color="skyblue")
plt.title("Los 10 Productos con Mayor Monto de Ventas")
plt.xlabel("Productos")
plt.ylabel("Monto Total (USD)")
plt.xticks(rotation=45, ha='right')  # Rotar etiquetas
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
